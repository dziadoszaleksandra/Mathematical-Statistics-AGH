```python
#Pytanie 2

#Średnia liczba wypadków, do których dochodzi na danym odcinku drogi wynosi 40 w ciągu roku.

#Oblicz prawdopodobieństwo, że w ciągu roku na danym odcinku drogi dojdzie do nie więcej niż 45 lub więcej niż 50 wypadków.


#Odpowiedź podaj w postaci ułamka z dokładnością do 4 miejsca po przecinku. Wynik zaokrąglij zgodnie z zasadami matematyki
```

```python
# Import potrzebnych modułów
from scipy.stats import poisson

# Ustawiamy wartość oczekiwaną λ
lambda_value = 40

# Krok 1: Obliczamy P(X <= 45)
prob_le_45 = poisson.cdf(45, lambda_value)  # Funkcja cdf liczy sumę prawdopodobieństw dla X = 0, 1, ..., 45

# Krok 2: Obliczamy P(X <= 50)
prob_le_50 = poisson.cdf(50, lambda_value)

# Krok 3: Obliczamy P(X > 50)
prob_gt_50 = 1 - prob_le_50

# Krok 4: Obie składowe są sumowane, bo zdarzenia są rozłączne:
result = prob_le_45 + prob_gt_50

# Krok 5: Wynik zaokrąglany jest do 4 miejsca po przecinku
result_rounded = round(result, 4)

print("Prawdopodobieństwo, że w ciągu roku wystąpi nie więcej niż 45 lub więcej niż 50 wypadków:", result_rounded)
```

```python
#Pytanie 3

#W urnie znajdują się kule białe (9) i czarne (6). Losujemy kulę 7 razy, za każdym razem wrzucając ją z powrotem do urny.

#Jakie jest prawdopodobieństwo, że biała kula zostanie wyciągnięta nie mniej niż 2, ale mniej niż 6 razy?


#Odpowiedź podaj w postaci ułamka z dokładnością do 4 miejsca po przecinku. Wynik zaokrąglij zgodnie z zasadami matematyki.
```

```python
import math

# Dane:
white = 9        # liczba białych kul
black = 6        # liczba czarnych kul
total = white + black  # łączna liczba kul: 15

# Prawdopodobieństwo wylosowania białej kuli
p = white / total   # 9/15 = 0.6
q = 1 - p           # 1 - 0.6 = 0.4

n = 7  # liczba losowań

# Inicjowanie zmiennej na sumę prawdopodobieństw dla k = 2,3,4,5
total_probability = 0

# Prawdopodobieństwo dla każdego k od 2 do 5
for k in range(2, 6):  # k = 2,3,4,5
    # Obliczamy symbol dwumianowy
    comb = math.comb(n, k)
    # Obliczamy prawdopodobieństwo dla dokładnie k białych kul
    p_k = comb * (p ** k) * (q ** (n - k))
    total_probability += p_k
    # Wyniki
    print(f"P(X = {k}) = {comb} * ({p:.4f})^{k} * ({q:.4f})^{n-k} = {p_k:.7f}")

# Łączne prawdopodobieństwo
print(f"\nŁączne prawdopodobieństwo (2 <= liczba białych kul < 6) = {total_probability:.7f}")

# Wynik zaokrąglony do 4 miejsc po przecinku
result_rounded = round(total_probability, 4)
print(f"Wynik (zaokrąglony do 4 miejsc po przecinku): {result_rounded}")
```

```python
#Pytanie 4

#Badania kliniczne wykazały, że prawdopodobieństwo wystąpienia komplikacji po wykonaniu pewnej operacji wynosi 5 %. 
#Oblicz prawdopodobieństwo, że w pewnym szpitalu mniej niż 6 lub nie mniej niż 13 operacji zakończy się sukcesem zanim po raz 3
#pojawią się pooperacyjne komplikacje.


#Odpowiedź podaj w postaci ułamka z dokładnością do 4 miejsca po przecinku. Wynik zaokrąglij zgodnie z zasadami matematyki.
```

```python
import math

# 1.Parametry zadania
p = 0.05            # Prawdopodobieństwo komplikacji (porażki)
q = 1 - p           # Prawdopodobieństwo sukcesu
r = 3               # Liczba komplikacji (porażeń), przy których zatrzymujemy się

# 2.Funkcja licząca prawdopodobieństwo dla rozkładu negatywno-dwumianowego:
# P(X=k) = C(k+r-1, k) * (q)^k * (p)^r, gdzie k to liczba sukcesów przed r porażkami.
def nb_pmf(k, r, p):
    return math.comb(k + r - 1, k) * (q ** k) * (p ** r)

# 3.Prawdopodobieństwo, że liczba sukcesów jest mniejsza niż 6 (k = 0, 1, 2, 3, 4, 5)
P_less_than_6 = sum(nb_pmf(k, r, p) for k in range(6))

# 4.Prawdopodobieństwo, że liczba sukcesów jest mniejsza lub równa 12 (k od 0 do 12)
# Wtedy prawdopodobieństwo, że sukcesów będzie co najmniej 13, wynosi:
P_at_least_13 = 1 - sum(nb_pmf(k, r, p) for k in range(13))

# 5.Łączne prawdopodobieństwo zdarzenia: mniej niż 6 lub co najmniej 13 sukcesów
wynik = P_less_than_6 + P_at_least_13

# 6.Wyświetlamy wynik zaokrąglony do 4 miejsc po przecinku
print(f"Wynik: {wynik:.4f}")
```
