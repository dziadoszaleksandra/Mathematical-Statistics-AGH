```python
#Pytanie 1

#Wzrost pewnej grupy ludzi podany w centymetrach jest dany rozkładem normalnym o wartości oczekiwanej 175i odchyleniu standardowym 8.
#Jakie jest prawdopodobieństwo, że losowo wybrana osoba z tej grupy będzie miała więcej niż 162, ale nie więcej niż 178 cm wzrostu?

#Odpowiedź podaj w postaci ułamka z dokładnością do 4 miejsca po przecinku. Wynik zaokrąglij zgodnie z zasadami matematyki.
```

```python
from scipy.stats import norm


# Parametry rozkładu normalnego
mu = 175      # wartość oczekiwana (średnia)
sigma = 8     # odchylenie standardowe


# Obliczenie prawdopodobieństwa, że wzrost jest większy niż 162 cm, a jednocześnie nie przekracza 178 cm
# norm.cdf(x, loc=mu, scale=sigma) oblicza dystrybuantę rozkładu normalnego dla wartości x.
p = norm.cdf(178, loc=mu, scale=sigma) - norm.cdf(162, loc=mu, scale=sigma)


# Zaokrąglenie wyniku do 4 miejsc po przecinku zgodnie z zasadami matematyki
p_rounded = round(p, 4)


print("Prawdopodobieństwo =", p_rounded)
```

```python
####Kod powoduje wyświetlenie prawdopodobieństwa, że losowo wybrana osoba z tej grupy ma wzrost większy
#niż 162 cm, ale nie przekraczający 178 cm, wyrażonego jako ułamek dziesiętny z dokładnością do 4 miejsc po przecinku.


####Wyjaśnienie kodu:
#1.Importowanie niezbędnej funkcji
from scipy.stats import norm
#Importujemy funkcję norm z modułu scipy.stats, która pozwala na korzystanie z funkcji związanych z rozkładem normalnym, takich jak dystrybuanta (CDF).

#2.Definicja parametrów rozkładu
mu = 175      # wartość oczekiwana (średnia)
sigma = 8     # odchylenie standardowe
#Ustalamy wartość oczekiwaną (średnią) μ=175 cm oraz odchylenie standardowe σ=8 cm.

#3.Obliczenie prawdopodobieństwa
p = norm.cdf(178, loc=mu, scale=sigma) - norm.cdf(162, loc=mu, scale=sigma)
#norm.cdf(178, loc=mu, scale=sigma) oblicza prawdopodobieństwo, że wzrost danej osoby jest mniejszy lub równy 178 cm.
#norm.cdf(162, loc=mu, scale=sigma) oblicza prawdopodobieństwo, że wzrost jest mniejszy lub równy 162 cm.
#Odejmując te dwie wartości, otrzymujemy prawdopodobieństwo, że wzrost mieści się w przedziale (162,178] cm. 
#W przypadku rozkładu ciągłego nie ma znaczenia, czy 162 jest włączone, czy nie.


#4.Zaokrąglenie wyniku
p_rounded = round(p, 4)
#Wynik jest zaokrąglany do 4 miejsc po przecinku zgodnie z zasadami matematycznymi.


#5.Wyświetlenie wyniku
print("Prawdopodobieństwo =", p_rounded)
#Program wypisuje końcowe, zaokrąglone prawdopodobieństwo .
```

```python
#Pytanie 2

#Funkcja gęstości prawdopodobieństwa średnicy D pewnego elementu schodzącego z linii produkcyjnej podanej w milimetrach dana jest wzorem:
#P(D=d)={3/4⋅(1−(d−100)^2)  dla 99≤d≤101
#       {0                  w pozostałych przypadkach
#Jakie jest prawdopodobieństwo, że losowo wybrany element będzie miał nie więcej niż 99.25 lub nie mniej niż 99.75 mm średnicy.

#Odpowiedź podaj w postaci ułamka z dokładnością do 4 miejsca po przecinku. Wynik zaokrąglij zgodnie z zasadami matematyki.
```

```python
import numpy as np
import scipy.integrate as integrate

# Definicja funkcji gęstości jako funkcji lambda
pdf = lambda d: 0.75 * (1 - (d - 100)**2) if 99 <= d <= 101 else 0.0

# Obliczenie P(D ≤ 99.25) = całka od 99 do 99.25
p_lower, _ = integrate.quad(pdf, 99, 99.25)

# Obliczenie P(D ≥ 99.75) = całka od 99.75 do 101
p_upper, _ = integrate.quad(pdf, 99.75, 101)

# Suma prawdopodobieństw
p = p_lower + p_upper

# Zaokrąglenie wyniku do 4 miejsc po przecinku
p_rounded = round(p, 4)

print("Prawdopodobieństwo =", p_rounded)
```

```python
###Kod oblicza łączne prawdopodobieństwo, że zmienna losowa D (o zadanej funkcji gęstości) przyjmie
#wartość mniejszą lub równą 99.25 lub większą lub równą 99.75 (w obrębie przedziału, w którym funkcja pdf jest różna od zera).




###Wyjaśnienie kodu:


#1.Importowanie bibliotek
import numpy as np
import scipy.integrate as integrate
#Importujemy bibliotekę NumPy (chociaż w tym kodzie nie jest ona używana) oraz moduł integrate z biblioteki scipy, który zawiera 
#funkcję do numerycznego całkowania.


#2.Definicja funkcji gęstości (pdf)
def pdf(d):
    if 99 <= d <= 101:
        return 0.75 * (1 - (d - 100)**2)  # 0.75 = 3/4
    else:
        return 0.0
#Funkcja pdf(d) definiuje gęstość prawdopodobieństwa dla zmiennej losowej D.
#Dla wartości d w przedziale [99,101] funkcja zwraca wartość 0.75⋅(1−(d−100)^2).
#Wartość 0.75 odpowiada 3/4.
#Jeśli d nie mieści się w przedziale [99,101], funkcja zwraca 0.0, co oznacza, że poza tym przedziałem gęstość jest zerowa.


#3.Obliczenie P(D≤99.25)
p_lower, _ = integrate.quad(pdf, 99, 99.25)
#Funkcja integrate.quad oblicza wartość całki oznaczonej funkcji pdf od d=99 do d=99.25.
#Wynik tej całki, czyli p_lower, odpowiada prawdopodobieństwu, że wartość D mieści się w przedziale [99,99.25].


#4.Obliczenie P(D≥99.75)
p_upper, _ = integrate.quad(pdf, 99.75, 101)
#Analogicznie, ta linia oblicza całkę funkcji pdf od d=99.75 do d=101.
#Wynik p_upper to prawdopodobieństwo, że D przyjmie wartość z przedziału [99.75,101].


#5.Sumowanie obliczonych prawdopodobieństw
p = p_lower + p_upper
#Suma p_lower+p_upper daje łączne prawdopodobieństwo, że wartość D znajduje się w przedziale od 99 do 99.25 lub od 99.75 do 101.


#6.Zaokrąglenie wyniku
p_rounded = round(p, 4)
#Funkcja round zaokrągla wynik do 4 miejsc po przecinku zgodnie z zasadami matematyki.


#7.Wyświetlenie wyniku
print("Prawdopodobieństwo =", p_rounded)
#Na końcu program wypisuje obliczone prawdopodobieństwo.
```

```python
#Pytanie 3

#Oczekiwany czas pracy linii produkcyjnej pomiędzy kolejnymi awariami wynosi 12 godzin.
#Jakie jest prawdopodobieństwo, że 3. awaria nastąpi w ciągu więcej niż 18, ale mniej niż 24 godzin od uruchomienia linii?

#Odpowiedź podaj w postaci ułamka z dokładnością do 4 miejsca po przecinku. Wynik zaokrąglij zgodnie z zasadami matematyki.
```

```python
import math


# Funkcja obliczająca dystrybuantę rozkładu Erlanga (gamma) dla k=3 i skali = 12
def gamma_cdf(t, k=3, scale=12):
    # Dla k = 3, dystrybuanta ma postać:
    # F(t) = 1 - exp(-t/scale) * (1 + t/scale + (t/scale)**2 / 2)
    x = t / scale
    return 1 - math.exp(-x) * (1 + x + x**2 / 2)


# Obliczamy wartości dystrybuanty dla t = 24 oraz t = 18 godzin
cdf_24 = gamma_cdf(24)
cdf_18 = gamma_cdf(18)


# Prawdopodobieństwo, że 3. awaria nastąpi między 18 a 24 godziną:
probability = cdf_24 - cdf_18


# Wynik zaokrąglony do 4 miejsca po przecinku
print(f"Prawdopodobieństwo: {probability:.4f}")
```

```python
####Kod wykorzystuje rozkład gamma (Erlanga) do obliczenia prawdopodobieństwa, że trzecia awaria wystąpi między 18 a 24 godziną, 
#przy założeniu, że średni czas między awariami wynosi 12 godzin.




####Wyjaśnienie kodu:


#1.Importowanie modułu:
import math
#Importujemy moduł math, który zawiera funkcje matematyczne, takie jak exp (funkcja wykładnicza).


#2.Definicja funkcji gamma_cdf:
def gamma_cdf(t, k=3, scale=12):
    x = t / scale
    return 1 - math.exp(-x) * (1 + x + x**2 / 2)
#Obliczenie dystrybuanty (CDF) rozkładu gamma (w tym przypadku rozkładu Erlanga) dla  k=3 oraz wartości oczekiwanej (skali) równiej 12.
#Parametry:t: punkt w czasie, dla którego liczymy dystrybuantę; k: liczba zdarzeń (tu 3 awarie). W naszym przypadku jest stała i wynosi 3; scale: średni czas między awariami, czyli 12 godzin.
#Obliczamy x= t/scale, czyli przeskalowany czas.
#Następnie korzystamy z wzoru na dystrybuantę rozkładu gamma (Erlanga) dla k=3

                                                                         
#3.Obliczanie wartości dystrybuanty dla określonych czasów:
cdf_24 = gamma_cdf(24)
cdf_18 = gamma_cdf(18)
#Wywołujemy funkcję gamma_cdf dla t=24 godzin, aby obliczyć prawdopodobieństwo, że 3. awaria nastąpi przed upływem 24 godzin.
#Analogicznie, obliczamy wartość dystrybuanty dla t=18 godzin.

      
#4.Obliczenie poszukiwanego prawdopodobieństwa:
probability = cdf_24 - cdf_18
#Różnica F(24)−F(18) daje prawdopodobieństwo, że 3. awaria wystąpi pomiędzy 18. a 24. godziną od uruchomienia linii produkcyjnej.

           
#5.Wypisanie wyniku:
print(f"Prawdopodobieństwo: {probability:.4f}")
#Wynik jest wypisywany z dokładnością do czterech miejsc po przecinku, zgodnie z zasadami matematycznego zaokrąglania.
```
